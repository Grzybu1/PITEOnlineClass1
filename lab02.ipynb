```python
import sklearn.datasets as sd
import pandas as pd
from sklearn.linear_model import LinearRegression as lr
import matplotlib.pyplot as pl
```

```python
rawData = sd.load_boston()
dataFrame = pd.DataFrame(rawData.data, columns = rawData.feature_names)
dataFrame.head()
```

```python
histogram = dataFrame.hist(column = ['AGE', 'DIS'], bins = 100)
```

```python
ageWithIndex = [(dataFrame.index[x],dataFrame['AGE'][x]) for x in dataFrame.index]
disWithIndex = [(dataFrame.index[x],dataFrame['DIS'][x]) for x in dataFrame.index]
regAge = lr().fit(ageWithIndex, rawData.target)
regDis = lr().fit(disWithIndex, rawData.target)
predictAge = regAge.predict(ageWithIndex)
predictDis = regDis.predict(disWithIndex)
```

```python
pl.clf()
pl.scatter(dataFrame.index,dataFrame['AGE'], 5)
pl.scatter(dataFrame.index,rawData.target, 5, 'red')
pl.plot(predictAge, 'green')
```

```python
pl.clf()
pl.scatter(dataFrame.index,dataFrame['DIS'], 2)
pl.scatter(dataFrame.index,rawData.target, 5, 'red')
pl.plot(predictDis, 'green')
```
